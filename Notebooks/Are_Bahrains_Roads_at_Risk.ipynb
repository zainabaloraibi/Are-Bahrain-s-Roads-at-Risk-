## Cleaning

### Drop Columns

```python
column_to_remove=['ROAD NAME','Unnamed: 1','ANALY-SIS','SECTION','Unnamed: 8','SURVEY DATE','EVENTS','DIRECTION','LANE NO.','SUB CHAINAGE']
raw_data1 = raw_data.drop(column_to_remove, axis=1)
```

```python
raw_data1.info()
```

### Change Columns Name

##### Change Rut Columns Name 

```python
raw_data1 = raw_data1.rename(columns={
    "RUT": "Rut Right",
    "Unnamed: 16": "Rut Left",
    "Unnamed: 17": "Rut Lane"
})
```

 ##### Change Texture Columns Name

```python
raw_data1 = raw_data1.rename(columns={
    "TEXTURE": "TEXTURE:SMTD  Right",
    "Unnamed: 19": "TEXTURE:SMTD  Left"
})
```

##### Change Ravelling Columns Name

```python
raw_data1 = raw_data1.rename(columns={
    "Unnamed: 36": "Road Porosity Index",
    "RAVELLING ":"Ravelling Index"
})
```

##### Change patholes Columns Name

```python
raw_data1 = raw_data1.rename(columns={
    "POTHOLES": "No. of Potholes",
    "Unnamed: 38": "Average Max Depth (mm)",
    "Unnamed: 39": "Severity"})
```

##### Change Distress Ratings Columns Name

```python
raw_data1 = raw_data1.rename(columns={
    "Distress Ratings": "IRI_rating",
    "Unnamed: 41": "Rut_rating",
    "Unnamed: 42": "Texture_rating",
"Unnamed: 43":"Crack_Rating",
"Unnamed: 44":"Ravelling_rating",
"Unnamed: 45":"Potholes_Rating"})
```

```python
##### Change Dominant Distress Columns Name
```

```python
raw_data1 = raw_data1.rename(columns={
    "Dominant Distress": "Min_rating",
    "Unnamed: 47": "rating"})
```

### Handle Nulls

```python
raw_data1.isnull().sum()
```

##### Check the nulls in Item No

```python
raw_data1.isnull().sum()
```

```python
raw_data1['IRI Avg'] = raw_data1['IRI Avg'].fillna('0')
```

```python
raw_data1.info()
```

##### Fill Severity

```python
raw_data1['Severity'].unique()
```

```python
severity_map = {
    5:'No severity',
    0:'No severity',
    4:'Low',
    3:'Low',
    2:'Moderate',
    1:'High'
}
raw_data1['Severity'] = raw_data1['Potholes_Rating'].map(severity_map).fillna('No severity')
```

```python
raw_data1['Severity'].unique()
```

```python
raw_data1=raw_data1.fillna(0)
```

```python
raw_data1.isnull().sum()
```

#### Drop location

```python
Lat_to_drop = raw_data1[raw_data1['LATITUDE'] == 'X'].index
raw_data1 = raw_data1.drop(Lat_to_drop) 
```

```python
raw_data1.info()
```

```python
raw_data1['in_bahrain'] = (
    (raw_data1['LATITUDE'] >= 25.7) & (raw_data1['LATITUDE'] <= 26.3) &
    (raw_data1['LONGITUDE'] >= 50.2) & (raw_data1['LONGITUDE'] <= 50.8)
)
```

```python
in_bahrain_drop = raw_data1[raw_data1['in_bahrain'] == False].index
raw_data1 = raw_data1.drop(in_bahrain_drop) 
```

```python
raw_data1.info()
```

```python
columns_with_x = raw_data1.columns[raw_data1.isin(['X']).any()]
print(columns_with_x)
```

```python
raw_data1[columns_with_x] = raw_data1[columns_with_x].replace("X", 0)
```

```python
Rut_to_drop = raw_data1[raw_data1['Rut Left'] == 'Rut Left (mm)'].index
raw_data1 = raw_data1.drop(Rut_to_drop) 
```

### Change types

```python
cols = ['LONGITUDE', 'LATITUDE','IRI Avg','Rut Right','Rut Left','Rut Lane','TEXTURE:SMTD  Right','TEXTURE:SMTD  Left','Road Porosity Index','No. of Potholes','Average Max Depth (mm)','Potholes_Rating','Ravelling_rating','Crack_Rating','Texture_rating','Rut_rating','IRI_rating']  
raw_data1[cols] = raw_data1[cols].astype('float64')
```

```python
raw_data1.info()
```

```python
raw_data1=raw_data1.reset_index(drop=True)
```

```python
raw_data1.info()
```

## Process Data

```python
process_data = raw_data1
```

##### Maximum Rut

```python
process_data['MAX_RUT']=process_data[['Rut Right','Rut Left','Rut Lane']].max(axis=1)
```

##### Average Texture

```python
process_data['AVG_TEXTURE']=process_data[['TEXTURE:SMTD  Right','TEXTURE:SMTD  Left']].mean(axis=1)
```

##### Cracks

```python
process_data['Zone1'] = (
    process_data['Zone1: Crack Intensity Over Cells'] *
    process_data['Zone1: Crack Width Over Cells'] *
    process_data['Zone1: Total Cracked Cells'] * 1000
) / (process_data['Zone1: Total Cells'] * 1000 * 1000 ) *100
```

```python
process_data.info()
```

```python
process_data['Zone2'] = (
    process_data['Zone2: Crack Intensity Over Cells'] *
    process_data['Zone2: Crack Width Over Cells'] *
    process_data['Zone2: Total Cracked Cells'] * 1000
) / (process_data['Zone2: Total Cells'] * 1000 * 1000 ) *100
```

```python
process_data['Zone3'] = (
    process_data['Zone3: Crack Intensity Over Cells'] *
    process_data['Zone3: Crack Width Over Cells'] *
    process_data['Zone3: Total Cracked Cells'] * 1000
) / (process_data['Zone3: Total Cells'] * 1000 * 1000 ) *100
```

```python
process_data['Cracks'] = process_data[['Zone1','Zone2','Zone3']].mean(axis=1)
```

###### Drop Columns in Zones that we dont need

```python
col_to_drop_zone = [
    'Zone1: Crack Intensity Over Segment',
    'Zone2: Crack Intensity Over Segment',
    'Zone3: Crack Intensity Over Segment'
]
process_data.drop(columns=col_to_drop_zone, inplace=True)
```

```python
Roads = process_data.groupby('ROAD NO.')['rating'].mean().reset_index()
```

```python
Roads.describe()
```

```python
def classify_road(avg_rating):
    if avg_rating <= 2.5:
        return "Poor"
    elif avg_rating <= 3.5:
        return "Moderate"
    elif avg_rating <= 4.5:
        return "Good"
    else:
        return "Very Good"

Roads['Rating_Category'] = Roads['rating'].apply(classify_road)
```
